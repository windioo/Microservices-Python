```python
import json


data_indonesia = open("covid_indonesia.json")
data_japan = open("covid_japan.json")
data_india = open("covid_india.json")


def read_data(key):
    data = {
        "Indonesia" : [],
        "Jepang" : [],
        "India" : []
    }

    buka_indonesia = json.load(data_indonesia)
    buka_japan = json.load(data_japan)
    buka_india = json.load(data_india)

    for r in range(len(buka_indonesia)):
        data["Indonesia"].append(buka_indonesia[r][key])
    for r in range(len(buka_japan)):
        data["Jepang"].append(buka_japan[r][key])
    for r in range(len(buka_india)):
        data["India"].append(buka_india[r][key])
    return data



covid = read_data("Cases")
print(covid)
```
